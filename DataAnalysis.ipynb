```python
def analysis(df,col,vartype,hue =None):    
    sns.set(style="darkgrid")
    if vartype == 0:
        fig, ax=plt.subplots(nrows =1,ncols=3,figsize=(20,8))
        ax[0].set_title("Distribution Plot")
        sns.distplot(df[col],ax=ax[0])
        ax[1].set_title("Violin Plot")
        sns.violinplot(data =df, x=col,ax=ax[1], inner="quartile")
        ax[2].set_title("Box Plot")
        sns.boxplot(data =df, x=col,ax=ax[2],orient='v')
    if vartype == 1:
        temp = pd.Series(data = hue,dtype=object)
        fig, ax = plt.subplots()
        width = len(df[col].unique()) + 10 + 10*len(temp.unique())
        fig.set_size_inches(width , 10)
        ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue) 
        if len(temp.unique()) > 0:
#             for p in ax.patches:
#                 ax.annotate('{:1.1f}%'.format((p.get_height()*100)/float(len(loan))), (p.get_x()+0.30, p.get_height()+20))  
#         else:
            for p in ax.patches:
                ax.annotate(p.get_height(), (p.get_x()+0.50, p.get_height()+2))
        del temp
    else:
        exit
    plt.show()
```

```python
analysis(df=data1,col='Brand',vartype=1)
```

```python
data1['Platform'].unique()
```

```python
brand_list=[]
platform_youtube=[]
platform_twitter=[]

for brand in data1['Brand'].unique():
    for platform in data1['Platform'].unique():
        if brand not in brand_list:
            brand_list.append(brand) 
        if (platform=='Twitter'):
            platform_twitter.append(len(data1[(data1['Brand']==brand) & (data1['Platform']=='Twitter')]))
        if (platform=='YouTube'):
            platform_youtube.append(len(data1[(data1['Brand']==brand) & (data1['Platform']=='YouTube')]))
        
        print(brand + ' brand for the platform ' + platform + ' is ',len(data1[(data1['Brand']==brand) & (data1['Platform']==platform)]))            
```

```python
list(zip(platform_youtube,platform_twitter))
```

```python
brand_list

df = pd.DataFrame({
    'Brand': brand_list,
    'Youtube': platform_youtube,
    'Twitter': platform_twitter
})
df
```

```python
df.plot(x="Brand", y=["Youtube", "Twitter"], kind="bar",figsize=(14,7))
```
